```python
import numpy as np
```

```python
#here is the program capable of generating a dataset

#this is just the initial random state/seed
np.random.seed(0)

#number of samples
m= 80

#we use a normal distribution to have values between 0 and 5
features_x = np.random.normal(0, 5, m)

#add the bias and create a column-row structure
X = np.column_stack((np.ones(m), features_x))

#divide into two classes
y = (features_x + np.random.normal(0, 2, m)) > 0
y = y.astype(int)
```

```python
#sigmoid function
def sigmoid(z):
  return 1 / (1+ np.exp(-z))

#here we are creatin the cost function with the gradient descent optimizator
def compute_cost(theta, X, y):
  m = y.size
  h = sigmoid(X.dot(theta)) #hypothesis
  cost = -np.sum(y * np.log(h) + (1-y) * np.log (1-h)) / m  #GD opt. formula
  return cost

#using the cost and sigmoid funct added with the learning rate alpha and the iterations to go through all the iterations m
# and creating a cost history to see the progres
def gradient_descent(X, y, theta, alpha, num_iters):
    m = y.size
    cost_history = []

    for i in range(num_iters):
        h = sigmoid(X.dot(theta)) #hypothesis
        theta -= alpha * (X.T.dot(h - y)) / m #theta optimizator
        cost_history.append(compute_cost(theta, X, y))
        probab = sigmoid(X.dot( theta))

    return theta, cost_history


alpha = 0.01
num_iters = 1000
#we initialize with theta at 0 so it wont have any previous knowledge, after some iterations it will create it's own bias
theta_initial = np.zeros(X.shape[1])

theta, cost_hist = gradient_descent(X, y, theta_initial, alpha, num_iters)

#print the final theta param and the cost history
print('this is the progresion of the parameters theta: ', theta)
print(f'this is the cost histroy: {cost_hist}')
```

```python
#using the sigmoid function and the now trained parameters we can do some predictions
def pred(X, theta):
  probab = sigmoid(X.dot(theta)) #here we use the sigmoid funtion and the new param
  return probab >= 0.5 #if the samples are more than 50% (1) they will be true

predictions = pred(X,theta_initial)

#print the predictions and acurracy
print(predictions)
print('acurracy: ', np.mean(predictions == y))
```
